# Lists

Lists describe arrays of data and are an example of an iterable data type. They are defined using square brackets and may contain objects of different types. For example, we can define a list as it follows:

````python
li = [-5, 7, 4, 9, 1, 12, 2]
````

````python
li
````

One of the most important operations done with a list or other iterable corresponds to indexing. _i.e.,_ Retrieve an object given its index position (Or key, as it we will see later). It is important that unlike MATLAB and like C/C++, Python lists are 0-indexed.

````python
li[0]
````

````python
li[3]
````

If we try to use an index that overflows a list, it will return an exception, instead of returning memory garbage unlike C/C++:

````python
li[20]
````

Lists can be modified, which means that is possible to change an index value:

````python
li[0] = -100
li
````

Another powerful operation is that of slicing, which allows to select a sublist using two indices. On Python, slicing is accomplished by using the notation ``[a:b]``. It is necessary to note that slicing is non-inclusive with respect to the last index. For instance, lets see some examples: 

````python
li[1:3]
````

````python
li[2:6]
````

Slicing works also with a single index, it may be the first or the second, as always ``:`` is present. On such cases the resulting list contains the elements from the starting/0 index up to the second/last index, depending if the missing index corresponds to the second or first, respectively:

````python
li[2:]
````

Selection from first up to fourth element:

````python
li[:4]
````

Slicing and indexing also works with negative numbers:

````python
li
````

````python
li[-1]
````

````python
li[-3:]
````

As syntactic sugar, we can invert a list using the following notation:

````python
li[::-1]
````

**Note:** On general, slicing notation takes the form ``[a:b:c]``, on which the resulting list elements correspond to those between ``a`` and ``b - 1`` on ``c`` steps. For instance, using this notation we can only take list elements that are on even indices:

````python
li[::2]
````

Finally, it is possible to observe that ``+`` and ``*`` can be used to concatenate and replicate, respectively:

````python
[0, 4, 7] + [2, 3]
````

````python
[0] * 4
````

## Ordering

To order a list/iterable elements, Python offers two built-in functions that can be used. The difference between both is related to its returning form after execution.

While ``sorted()`` function returns a new list that contains the sorted elements of the original list, the ``list.sort()`` function sorts the list elements in-place, modifying the input list. For example:

````python
example_list = [-1, 0, 100, -29, 2, 3, 6]
````

First, we will obtain a new list using ``sorted``

````python
ordered_list = sorted(example_list)
ordered_list
````

Now, we will sort the list in-place, modifying its original contents:

````python
example_list.sort()
example_list
````

# Iterable unpacking

On Python we can "unpack" an iterable contents and assign them to variables using the notation ``var1, var2, ..., varN = iterable``. To achieve this, it is necessary that the iterable object contains ``N`` elements, otherwise it will raise an exception:

````python
# Unpack a list
a, b, c = [2, 3, 4]
print(a, b, c)
````

````python
# Unpack a string
a, b, c = '123'
print(a, b, c)
````

````python
# Unpack a tuple
a, b, c = 4, -2, 1+9j
print(a, b, c)
````

````python
# Less elements than variables:
a, b, c = [2, 3]
````

````python
# Less variables than elements
a, b = [4, 5, 6] 
````

## Funciones internas

Python permite definir funciones al interior de otra funci�n, las cuales se denominan funciones internas. Estas funciones pueden acceder al espacio de variables de la funci�n que la encapsula. Sin embargo, estas funciones no son accesibles desde el exterior de la funci�n.

Por ejemplo, a continuaci�n se presenta una funci�n que toma por p�rametro un elemento y retorna una lista con distintas operaciones sobre el elemento inicial:

````python
def doblar_y_dividir_por_dos(x):
    def doblar():
        return 2 * x

    def dividir_por_dos():
        return x / 2

    return [doblar(), dividir_por_dos()]
````

## M�todos

Como ya dijimos, los m�todos son funciones asociadas a una clase que operan sobre sus atributos. Por ejemplo, a la clase anterior le podemos a�adir un m�todo que calcule el m�dulo de un n�mero complejo con la f�rmula:

$$\left| z \right|=\sqrt{\textrm{Re}\left(z\right)^{2}+\textrm{Im}\left(z\right)^{2}}$$

Para ello redefinimos `NumeroComplejo` para agregarle un nuevo m�todo `modulo`, as�:

````python
class NumeroComplejo:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def modulo(self):
        return (self.real**2 + self.img**2)**(1/2)
````

**Nota**:

Al igual que para `__init__`, el primer argumento de todo m�todo debe ser `self`, para indicar que hace parte de la clase.

Con esta nueva definici�n obtenemos el siguiente resultado para el m�dulo del n�mero complejo definido a continuaci�n:

````python
z = NumeroComplejo(1, 2)
````

````python
z.modulo()
````

Aqu� puede parecer un poco extra�o que `modulo` se llame sin argumentos, cuando al definirlo en la clase se le hab�a pasado a `self` como primer argumento. Esto se debe a que `self` no es un argumento en s�, sino que s�lo se usa para se�alar que una funci�n es un m�todo de la clase, como ya se mencion�.

Otra operaci�n que puede hacerse con n�meros complejos es obtener su *conjugado*. El conjugado de un complejo $z$, es un nuevo n�mero complejo que se denota $\bar{z}$ y se define como

$$z=a+ib \longrightarrow \bar{z}=a-ib$$

Para obtener el conjugado podemos entonces agregar un nuevo m�todo a nuestra clase, de la siguiente forma:

````python
class NumeroComplejo(object):
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def modulo(self):
        return (self.real**2 + self.img**2)**(1/2)

    def conjugado(self):
        return NumeroComplejo(self.real, -self.img)
````

Para calcular el conjugado de un n�mero complejo `z`, s�lo debemos llamar el m�todo correspondiente y asign�serlo a una nueva variable `z1` (en caso de que deseemos usarlo m�s tarde):

````python
z = NumeroComplejo(1, 2)
````

Finalmente, vamos a a�adir un m�todo que retorne el producto de dos n�meros complejos. Dados dos n�meros

$$
z = a + ib \\
w = c + id
$$

su producto est� dado por:

$$z \times w = (ac - bd) + i(ad + bc)$$

Para ello podemos escribir el siguiente m�todo, llamado `producto`, en nuestra clase:

````python
class NumeroComplejo(object):
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def modulo(self):
        return (self.real**2 + self.img**2)**(1/2)

    def conjugado(self):
        return NumeroComplejo(self.real, -self.img)

    def producto(self, w):
        real = self.real * w.real - self.img * w.img
        img = self.real * w.img + self.img * w.real
        return NumeroComplejo(real, img)
````

````python
z = NumeroComplejo(1, 2)
w = NumeroComplejo(4, -7)
````

````python
x = z.producto(w)
````

````python
x.real
````

````python
x.img
````

Para comprobar que `producto` est� funcionando correctamente podemos usar la siguiente f�rmula, que relaciona el m�dulo de un n�mero complejo con su conjugado:

$$\left| z \right| = \sqrt{\textrm{Re} \left( z \times \bar{z} \right)}$$

````python
z2 = z.producto(z.conjugado())
````

````python
(z2.real)**(1/2) == z.modulo()
````
